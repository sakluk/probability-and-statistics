```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rnd
%matplotlib inline
```

```python
# Help for randn
rnd.randn?
```

```python
# Creating normally distributed random numbers
N = 10000
x = rnd.randn(10000)
p1 = plt.hist(x, 20)
```

```python
# How many of the values are below 0.0?
tf = x < 0.0
s = np.sum(tf)
p = s/N
print("s = ", s, "p = ", p)
```

```python
# How many of the values are between -1.0 and +1.0?

# We need to use logical_and function
tf = np.logical_and(-1.0 < x, x < 1.0)

s = np.sum(tf)
p = np.sum(tf)/N
print("s = ", s, "p = ", p)
```

```python
# Can we visualise it?
b = np.arange(-4, 4, 0.1)
plt.hist(x, b);
plt.hist(x[tf], b);
plt.title('Values between -1.0 and +1.0')
```

```python
# Random numbers having mean = 10 and std = 4
m = 10.0
sd = 4.0
x = rnd.randn(N)*sd + m
plt.hist(x, 20);
print("mean(x) = ", np.mean(x))
print("std(x) = ", np.std(x))
```

```python
# More functions from scipy.stats
# http://docs.scipy.org/doc/scipy/reference/tutorial/stats.html
from scipy import stats
from scipy.stats import norm
```

```python
# Plot the normal distribution
x = np.arange(-5, 5, 0.1)
y = norm.pdf(x)
plt.plot(x,y)
```

```python
# Plot the cumulative distribution
y = norm.cdf(x)
plt.plot(x,y)
```

```python
# Mean, std and variance
norm.mean(), norm.std(), norm.var()
```

```python
# What is the probability that x < 0.0?
p = norm.cdf(0.0)
print("P(x<0.0) = ", p)
```

```python
# How to visualize it?
x = np.arange(-4, 4, 0.1)
y = norm.pdf(x)
plt.fill_between(x, 0, y)
x2 = np.arange(-4, 0.1, 0.1)
y2 = norm.pdf(x2)
plt.fill_between(x2, 0, y2, facecolor = "green")
```

```python
# What is the probability that -1.0 < x < 1.0?
p = norm.cdf(1.0) - norm.cdf(-1.0)
print("P(-1.0 < x < 1.0) = ", p)
```
